# NUMPY

#### NUMPY ARRAYS AS COMPARED TO LISTS IN PYTHON, TAKES SIGNIFICANTLY MUCH LESSER STORAGE SPACE AND COMPUTATION TIME

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### declaring numpy array

```python
a = np.array([1, 2, 3])
a = np.array([[1, 2, 3], [4, 5, 6]])
a
```

#### arange function 

```python
a = np.arange(25)
a
```

#### get the dimension of array

```python
a.ndim
```

#### get size of element in the array in bytes

```python
a.itemsize
```

#### datatype of array

```python
a.dtype
```

#### size of array(total no. of elements in array)

```python
a.size
```

#### shape of array

```python
a.shape
```

#### reshaping array

```python
#reshape(row, col)
a = np.array([[1, 2, 3, 4], [3, 4, 5, 6]])
a = a.reshape(4, 2)
a
```

#### slicing array

```python
#a[row1:row2, col1:col2]
a = np.array([[1, 2, 3, 4], [3, 4, 5, 6], [5, 6, 7, 8]])
a[0:2, 1:3]
```

#### equally seperated values between limits

```python
#linspace(start limit inclusive, end limit inclusive, steps)
a = np.linspace(1, 3, 10)
a
```

#### vectorized sum

```python
a = np.array([[1, 2, 3], [4, 5, 6]])
b = np.array([[1, 2, 3], [4, 5, 6]])
a+b
```

#### vectorized subtraction

```python
a-b
```

#### vectorized multiplication

```python
a*b
```

#### vectorized division

```python
a/b
```

#### concatinating arrays

```python
#vertical concatination
np.vstack((a, b))
```

```python
#horizontal concatination
np.hstack((a, b))
```

#### converting multi-dim array to a single column

```python
a.ravel()
```

#### maximum element in array

```python
a.max()
```

#### minimum element in array

```python
a.min()
```

#### sum of all elements in array

```python
a.sum()
```

#### getting sum of each column as a list

```python
#for columns, axis = 0
a.sum(axis=0)
```

#### getting sum of each row as a list

```python
#for rows, axis = 1
a.sum(axis=1)
```

#### sqr root

```python
np.sqrt(a)
```

#### standard deviation

```python
np.std(a)
```

#### trignometric functions

```python
x = np.arange(0, 3*np.pi, 0.1)
```

```python
#sine function
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
#cosine function
y = np.cos(x)
plt.plot(x, y)
plt.show()
```

```python
#tan function
y = np.tan(x)
plt.plot(x, y)
plt.show()
```

#### exponential, log functions

```python
a = np.array([1, 2, 3])
np.exp(a)
```

```python
np.log(a)
```

```python
np.log10(a)
```
