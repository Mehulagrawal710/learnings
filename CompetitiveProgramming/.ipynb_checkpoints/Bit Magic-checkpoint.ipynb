## check if a number is power of 2

```python
num = 16
if num & (num-1) == 0:
    print(True)
else:
    print(False)
```

## Set a bit at nth lsb position in number

```python
"""use OR operator"""
num = 23 # binary=> 10111
nlsb = 4
num = num | 1<<(nlsb-1)
print(num, bin(num))
```

## UnSet a bit at nth lsb position in number

```python
"""use AND operator"""
num = 15 # binary - 1111
nlsb = 3
num = num & ~(1<<(nlsb-1))
print(num, bin(num))
```

## Toggle a bit at nth lsb position in number

```python
"""use XOR operator"""
num = 15 # binary - 1111
nlsb = 3
num = num ^ (1<<(nlsb-1))
print(num, bin(num))
```

## check a bit at nth lsb position in number

```python
"""use AND operator"""
num = 17 # binary - 10001
nlsb = 5
res = num & (1<<(nlsb-1))
if res==0:
    print(0)
else:
    print(1)
```

## Flipping bits of a number / 1s-complement

```python
import math
def flipBits(num, nbits):
    # nbits represent the number of bits to consider, eg- to deal with 32 bit integer, pass nbits = 32
    for i in range(nbits):
        num = num^(1<<i)
    return num

"""
5 - 101
2 - 010
"""
num = 5
#nbits = math.floor(math.log(num, 2))+1
nbits = 8
flipped = flipBits(num, nbits)
print("number => ", num, "-", bin(num)[2:])
print("flipped {}-bit binary => ".format(nbits), flipped, "-", bin(flipped)[2:])
```

## 2s-complement

```python
def twoscomplement(num, nbits):
    i = 0
    while num & 1<<i==0:
        # while current bit is 0
        i += 1
    i += 1 # previous bit was 1
    # flipping remaining bits
    for j in range(i, nbits):
        num = num ^ (1<<j)
    return num

num = 20 # 10100
nbits = 8
com2 = twoscomplement(num, nbits)
print(com2, bin(com2))
```

```python
"""
Solution:
Find the remainder of N by moduling it with 4.
If rem = 0, then xor will be same as N.
If rem = 1, then xor will be 1.
If rem = 2, then xor will be N+1.
If rem = 3 ,then xor will be 0.

Number Binary-Repr  XOR-from-1-to-n
1         1           [0001]
2        10           [0011]
3        11           [0000]  <----- We get a 0
4       100           [0100]  <----- Equals to n
5       101           [0001]
6       110           [0111]
7       111           [0000]  <----- We get 0
8      1000           [1000]  <----- Equals to n
9      1001           [0001]
10     1010           [1011]
11     1011           [0000] <------ We get 0
12     1100           [1100] <------ Equals to n
"""
n = 6
if n%4==0: print(n)
elif n%4==1: print(1)
elif n%4==2: print(n+1)
elif n%4==3: print(0)
```
